```python
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
import surprise
```

```python
os.chdir('C:/Users/Shivam/Downloads/ml-100k/ml-100k')
```

```python
data=Dataset.load_builtin('ml-100k')
trainset,testset= train_test_split(data,test_size=.15)   #15%
```

```python
#use user based true/false to switch between user-based or item base collaborative
```

```python
algo=KNNWithMeans(k=50,sim_options={'name':'pearson_baseline' ,  'user_based':True}) #50 nearest neighbour
algo.fit(trainset)
```

```python
uid= str(196)  #raw user id
iid=str(302) #raw item id
```

```python
#get a prediction for specific users and items

pred= algo.predict(uid,iid,verbose=True)
```

```python
#run the trained model against testset
test_pred=algo.test(testset)
```

```python
test_pred
```

```python
#GET rmse   
print("user-based Model:Test Set")

accuracy.rmse(test_pred,verbose=True)
```
